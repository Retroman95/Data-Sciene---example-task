```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
path = 'https://github.com/anastasiarazb/skillbox-data-analyst-intensive/blob/master/%D0%90%D0%BD%D0%B0%D0%BB%D0%B8%D0%B7%20%D0%B4%D0%B0%D0%BD%D0%BD%D1%8B%D1%85%20%D0%94%D0%B5%D0%BD%D1%8C%202/IPG2211A2N.xls?raw=true'

data = pd.read_excel(path, skiprows=10)

data
```

```python
data.columns
```

```python
x = data['observation_date']
y = data['IPG2211A2N']
plt.figure(figsize=(20, 5))
plt.plot(x, y, label='IPG2211A2N - данные из файла (употреблённая энергия)')
plt.title('Полученные данные')
plt.legend()
```

```python
data['min_IPG2211A2N'] = data['IPG2211A2N'].min()
data['IPG2211A2N'].min()
```

```python
data['max_IPG2211A2N'] = data['IPG2211A2N'].max()
data['IPG2211A2N'].max()
```

```python
x = data['observation_date']
y = data['IPG2211A2N']
plt.figure(figsize=(20, 5))
plt.plot(x, y, label='IPG2211A2N - данные из файла (употреблённая энергия)')
plt.plot(x, data['min_IPG2211A2N'], label='min')
plt.plot(x, data['max_IPG2211A2N'], label='max')
plt.legend()
```

```python
data['quantile_95'] = data['IPG2211A2N'].quantile(0.95)
data['quantile_05'] = data['IPG2211A2N'].quantile(0.05)
```

```python
x = data['observation_date']
y = data['IPG2211A2N']
plt.figure(figsize=(20, 5))
plt.plot(x, y, label='IPG2211A2N - данные из файла (употреблённая энергия)')
plt.plot(x, data['min_IPG2211A2N'], label='min')
plt.plot(x, data['max_IPG2211A2N'], label='max')
plt.plot(x, data['quantile_95'], label='quantile_95')
plt.plot(x, data['quantile_05'], label='quantile_05')
plt.legend()
```

```python
data['IPG2211A2N'].rolling(10).mean()
```

```python
x = data['observation_date']
y = data['IPG2211A2N']
plt.figure(figsize=(20, 5))
plt.plot(x, y, label='IPG2211A2N - данные из файла (употреблённая энергия)')
plt.plot(x, data['min_IPG2211A2N'], label='min')
plt.plot(x, data['max_IPG2211A2N'], label='max')
plt.plot(x, data['quantile_95'], label='quantile_95')
plt.plot(x, data['quantile_05'], label='quantile_05')
plt.plot(x, data['IPG2211A2N'].rolling(10).mean(), label='Скользящее среднее 10 месяцев')
plt.plot(x, data['IPG2211A2N'].rolling(100).mean(), label='Скользящее среднее 100 месяцев')
plt.plot(x, data['IPG2211A2N'].rolling(200).mean(), label='Скользящее среднее 200 месяцев')
plt.legend()
```

```python
# Создаем объект линейного регрессора
model = LinearRegression()
# Обучаем модель на точках
X = pd.DataFrame(data.index)  # преобразовали индекс в таблицу
model.fit(X, data['IPG2211A2N'])  # Обучаемся на исходных данных
```

```python
result = model.predict(X)  # "предсказали" выровненные точки
```

```python
x = data['observation_date']
plt.plot(x, data['IPG2211A2N'])
plt.plot(x, result)
```
